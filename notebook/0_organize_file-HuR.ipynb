```python
import os
indir = '/projects/ps-yeolab5/hsher/Tao_circSTAMP'

fastqs = []
libs = []
for folder in os.listdir(indir):
    if folder !='fastq':
        if os.path.isdir(os.path.join(indir, folder)):
            files = os.listdir(os.path.join(indir, folder))
            for file in files:
                if file.endswith('fastq.gz'):
                    libs.append(file.split('_')[0])
                    fastqs.append(os.path.join(indir, folder, file))
```

```python
import pandas as pd
data = pd.DataFrame([libs, fastqs], index = ['lib', 'fastq_path']).T
```

```python
data['read'] = data['fastq_path'].apply(lambda file: '1' if 'R1' in file else '2')
```

```python
data.groupby(by = ['lib', 'read']).count()
```

```python
outdir='/projects/ps-yeolab5/hsher/Tao_circSTAMP/fastq/'
for name, group in data.groupby(by = ['lib', 'read']):
    libname = name[0]
    readn = name[1]
    all_files = ' '.join(group['fastq_path'].tolist())
    os.system(f'cat {all_files} > {outdir}{libname}.{readn}.fastq.gz')
    print(f'cat {all_files} > {outdir}{libname}.{readn}.fastq.gz')
```

# make menifest

```python
libs = list(set([f.split('.')[0] for f in os.listdir(outdir)]))
fastq1 = [os.path.join(outdir,f'{l}.1.fastq.gz') for l in libs]
fastq2 = [os.path.join(outdir,f'{l}.2.fastq.gz') for l in libs]
menifest = pd.DataFrame([libs, fastq1, fastq2], index = ['Sample', 'fastq1', 'fastq2']).T
```

```python
menifest['Rnase'] = menifest['Sample'].apply(lambda name: True if name.endswith('-R') else False)
```
