````python
from pathlib import Path
````

````python
outdir = Path('fastq_csv')
import os
try:
    os.mkdir(outdir)
except Exception as e:
    print(e)
    pass

def check_manifest(df, fname):
    assert not df['Sample'].duplicated().any()
    assert not df['fastq1'].duplicated().any()
    assert not df['fastq2'].duplicated().any()
    assert df['fastq1'].apply(lambda f: Path(f).exists()).all()
    assert df['fastq2'].apply(lambda f: Path(f).exists()).all()
    df.to_csv(outdir / f'{fname}.csv')
````
